## Importing Libraries

```python
%matplotlib inline
import cv2 
import numpy as np 
from matplotlib import pyplot as plt 
from matplotlib import image as image
from matplotlib import pyplot as plt
import easygui 
```

## Read and Displaying Image with the help of imshow() function
Also, converted the image to grayscale

```python
I = cv2.imread("Faded.jpg",0)
plt.imshow(  I,cmap='gray')
```

## Histogram equalization
Used to increase the contrast of the image or poor lighting

```python
equilized_img = cv2.equalizeHist(I)
plt.imshow(equilized_img,cmap = 'gray')
```

## Canny edge detection 
Used to detect wide range of edges in the image

Applying the canny edge detection on original and Normalized image and also displaying both of them on screen using subplot function

canny1 = Used for original image

canny2 = Used for normalised image

Note : while displaying the image all are set to 'gray' as a grayscale



```python
canny1 = cv2.Canny(I,100,200)
canny2 = cv2.Canny(equilized_img,100,200)
plt.subplot(2,2,1),plt.title('Original image after Canny')
plt.imshow(canny1,cmap='gray')
plt.subplot(2,2,2),plt.title('normalized image after Canny')
plt.imshow(canny2,cmap='gray')
```

## Sobel detection on X- axis and Y - axis
Used this approach to detect the edges of the images in both horizontal and virtical directions

Sobel1 = detection on X-axis

Sobel2 = detection on Y-axis

Its a really good way to differentiate out both axis as we can clearly see on the Y-axis it, completly remove the faded area or can say the removed the edges of inner rectangle.

```python
sobel1 = cv2.Sobel(I,cv2.CV_64F,1,0)
plt.subplot(2,2,1),plt.title('Sobel on X-axis')
plt.imshow(sobel1,cmap='gray')

sobel2 = cv2.Sobel(equilized_img,cv2.CV_64F,0,1)
plt.subplot(2,2,2),plt.title('Sobel on Y-axis')
plt.imshow(sobel2,cmap='gray')
```

## Image Masking

```python
ret,mask = cv2.threshold(I, 170,255,cv2.THRESH_BINARY_INV)
test = cv2.bitwise_and(I,I, mask=mask)
plt.imshow(mask,cmap='gray')
```
